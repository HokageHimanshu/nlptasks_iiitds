```python
import pandas as pd
import numpy as np
import string
import datetime

from oauth2client.service_account import ServiceAccountCredentials
import gspread
from google.oauth2.service_account import Credentials
from googleapiclient.discovery import build
```

```python
class SheetEditor():
        def __init__(self,json_file):
            self.headers = ['ID','WORD','LEMMA','UPOS','XPOS','FEATS','HEAD','DEPREL','DEPS','MISC']
            scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
            self.creds = ServiceAccountCredentials.from_json_keyfile_name(json_file, scope) 
            self.client = gspread.authorize(self.creds)
        
        def readEntireSheet(self, sheet_instance=None):
            data = sheet_instance.get_all_values()
            df = pd.DataFrame(data[1:],columns=self.headers)
            return df
        
        def readSheetByRange(self, sheet_instance=None, data_range='A2:J', header_range='A1:J1'):
#             data = sheet_instance.get_values(header_range)
            data1 = sheet_instance.get_values(data_range)
            df = pd.DataFrame(data1,columns=self.headers)
            return df
        
#         def writeSheetFromIndex(self, sheet_instance=None, data_df=None, header_row=1, data_row=2, add_header=False):    
#             if add_header:
#                 sheet_instance.insert_row(data_df.columns.tolist(),header_row)
#                 sheet_instance.freeze(header_row)
#             sheet_instance.insert_rows(data_df.values.tolist(),data_row)
#         def deleteRowsFromeSheet(self,sheet_instance,start_row=1,end_row=40):
#             sheet_instance.delete_rows(start_row,end_row)
```

```python
# configerations to connect to sheet

json_file_path = 'tamil-nlp-4f5946578e44.json'
sheet_editor = SheetEditor(json_file_path)
```

```python
# configure them according to the sheet

gsheets_name = "Team 1 - Real Dataset Annotation"
sub_sheet_name = "sentences_to_be_annotated"
data_range = 'A2:J2000'
```

```python
sheet = sheet_editor.client.open(gsheets_name)
sheet_instance = sheet.worksheet(sub_sheet_name)
records_df = sheet_editor.readSheetByRange(sheet_instance, data_range=data_range)
print('Successfully read the data')
```

```python
# sample code of traversing the data and perform

def sample_func_to_count_sentences(df):
    count=0
    l = []
    for i in range(len(df)):
        row = df.iloc[i]
        if 'sent_id' in row['ID']:
            count+=1
            l.append(row['ID'].split('=')[1])
    return count,l

count, ids = sample_func_to_count_sentences(records_df)
print('The number of sentences in the choose range are ', count)
print('They are the following = ', '\n'.join(ids))
```
